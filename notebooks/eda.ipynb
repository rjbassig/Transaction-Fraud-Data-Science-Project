```python
#Set up
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Data Frame Print and Info
df = pd.read_csv("../Credit Data/creditcard.csv")
print(df.head())
print(df.info())
```

```python
#Amunt of Fraud vs No Fraud
df['Class'].value_counts()
```

```python
#Class Distribution 
sns.countplot(x="Class", data=df)
plt.title("Fraud vs. Non-Fraud")
plt.show()
```

```python
sns.histplot(df["Amount"], bins = 50)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
#Log Transform
df['Log Amount'] = np.log1p(df["Amount"])

sns.histplot(df["Log Amount"], bins=50)
plt.title("Log Amount Distribution")
plt.show()
```

```python
#Fraud Vs Amount
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Amount by Fraud vs Non-Fraud")
plt.show()
```
